## Parte B
Calcular de manera numéricamente las siguientes EDO's, usando valores iniciales y rangos apropiados.
1. **Resolver (Hint: separable)**
$$
\frac{dy}{dx}=x+xy.
$$

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def ode(y,x):
    dydx = x+x*y
    return dydx
t_vec = np.linspace(0,1,101)
y_ini = (1,0)
f = odeint(ode, y_ini, t_vec)
plt.plot(t_vec, f[:,0])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

2. **Resolver (Hint: exacta)**
$$
x\frac{dy}{dx}+3x+y=0.
$$

```python
def ode(y, x):
    dydx = -3-y/x
    return dydx
t_vec = np.linspace(0.1,1,101)
y_ini = (1,0)
f = odeint(ode, y_ini, t_vec)
plt.plot(t_vec, f[:,0])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

3. **Resolver (hint: μ)**
$$
\frac{dy}{dx}+2xy=4x.
$$

```python
def ode(y, x):
    dydx = 4*x-2*x*y
    return dydx
t_vec = np.linspace(0,1,101)
y_ini = (1,0)
f = odeint(ode, y_ini, t_vec)
plt.plot(t_vec, f[:,0])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

4. **Resolver (Hint: homogenea)**
$$
\frac{dy}{dx}=\frac{y}{x}+\tan(\frac{y}{x}).
$$

```python
def ode( y, x):
    dydx = y/x + np.tan(y/x)
    return dydx
t_vec = np.linspace(0.1,1,101)
y_ini = (np.pi,0)
f = odeint(ode, y_ini, t_vec)
plt.plot(t_vec, f[:,0])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

5. **Resolver (Hint: Bernoulli)**
$$
\frac{dy}{dx}+\frac{y}{x}=2x^3y^4.
$$

```python
def ode(y, x):
    dydx = 2*x**3*y**4-y/x
    return dydx
t_vec = np.linspace(0.1,1,101)
y_ini = (1,0)
f = odeint(ode, y_ini, t_vec)
plt.plot(t_vec, f[:,0])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

6. **Resolver (Hint: $y_p+y_c$)**
$$
\frac{d^2y}{dx^2}-2\frac{dy}{dx}+y=e^x,
$$
$$
x^2\frac{d^2y}{dx^2}+x\frac{dy}{dx}-4y=0.
$$

```python
def ode(Y, x):
    y, v = Y
    dydx = [v, 2*v-y +np.exp(x)]
    return dydx
t_vec = np.linspace(0,5,101)
y_ini = (np.pi,0)
f = odeint(ode, y_ini, t_vec)
plt.plot(t_vec, f[:,0])
plt.plot(t_vec, f[:,1])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def ode(Y, x):
    y, v = Y
    dydx = [v, -1/x*v+4*y/x**2]
    return dydx
t_vec = np.linspace(0.1,5,101)
y_ini = (1,0)
f = odeint(ode, y_ini, t_vec)
plt.plot(t_vec, f[:,0])
plt.plot(t_vec, f[:,1])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import optimize,  interpolate, integrate
from scipy.integrate import quad
import seaborn as sns
import numpy as np
import pandas as pd

%matplotlib inline
```

Usar SEABORN para graficar

```python
sns.set()
```

Leer archivos con PANDAS y almacenar

```python
pdf = pd.read_csv('file_f.csv', delimiter=',')
pdf.head()
```

```python
pdg = pd.read_csv('file_g.csv', delimiter=',')
pdg.head()
```

Graficar archivos

```python
sns.scatterplot(x='x', y='y', data=pdf, c='red', marker='o', s=10)
sns.scatterplot(x='x', y='y', data=pdg, c='green', marker='+')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Sumar componentes y's y graficar

```python
pds = pdf.copy()
pds['y'] = pdf['y'] + pdg['y']
sns.scatterplot(x='x', y='y', data=pds, c='blue', edgecolor="white")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Ajustar una funcion cuadratica y encontrar coeficientes

```python
def f(t, a, b, c):
    return a*t**2 + b*t + c

popt, pcov = optimize.curve_fit(f, pds['x'], pds['y'])
a, b, c = popt
print(f"Coeficientes de la función cuadrática ajustada:\na = {a}\nb = {b}\nc = {c}")
```

Graficar promedios y matriz de covarianza

```python
x_mean = np.mean(pds['x'])
y_mean = np.mean(pds['y'])

print(f"El promedio de x: {x_mean}")
print(f"El promedio de y: {y_mean}")

print("Matriz de Covarianza:")
print(pcov)

plt.figure(figsize=(8, 6))
sns.heatmap(pcov, annot=True, cmap='coolwarm', cbar=True, xticklabels=['a', 'b', 'c'], yticklabels=['a', 'b', 'c'])
plt.show()
```

Graficar mejor ajuste de la cuadratica

```python
sns.scatterplot(x='x', y='y', data=pds, c='blue', edgecolor="white")
plt.plot(pds['x'], f(pds['x'], *popt), 'r-')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

El mejor ajuste convertir a una función

Encontrar raíces

```python
def f_x(x):
    return a*x**2 + b*x + c

root_1 = optimize.root_scalar(f_x, bracket=[-1, 1], method='brentq')
root_2 = optimize.root_scalar(f_x, bracket=[1, 3], method='brentq')
print(root_1, "\n")
print(root_2)
```

Encontrar minimo

```python
min = optimize.minimize(f_x, x0=1)
min
```

Calcular derivadas

Graficar tangentes a raices y funcion

```python
def d_f(x):
    return 2*a*x + b

def l_tan(x, x_0):
    return d_f(x_0)*(x-x_0) + f_x(x_0)

x_1 = np.linspace(-2, 1.3, 100)
x_2 = np.linspace(0.5, 4, 100)

sns.lineplot(x=x_1, y=l_tan(x_1, root_1.root), color='green')
sns.lineplot(x=x_2, y=l_tan(x_2, root_2.root), color='red')
plt.plot(pds['x'], f(pds['x'], *popt), 'b-')
plt.annotate(f'({root_1["root"]:.2f}, 0)', (-0.1, 0), textcoords="offset points", xytext=(0,10), ha='center')
plt.annotate(f'({root_2["root"]:.2f}, 0)', (3.1, 0), textcoords="offset points", xytext=(0,10), ha='center')
plt.annotate(f'({min["x"][0]:.2f}, {f_x(min["x"][0]):.2f})', (1.5, -13), textcoords="offset points", xytext=(0,10), ha='center')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Graficar area entre parabola y tangentes

```python
x_1 = np.linspace(root_1.root, min["x"][0], 100)
x_2 = np.linspace(root_2.root, min["x"][0], 100)
x_3 = np.linspace(root_1.root, root_2.root, 100)
sns.lineplot(x=x_1, y=l_tan(x_1, root_1.root), color='green')
sns.lineplot(x=x_2, y=l_tan(x_2, root_2.root), color='red')
sns.lineplot(x=x_3, y=f_x(x_3), color='blue')

plt.fill_between(x_1, f_x(x_1), l_tan(x_1, root_1.root), color='lightgreen', alpha=0.5)
plt.fill_between(x_2, f_x(x_2), l_tan(x_2, root_2.root), color='lightcoral', alpha=0.5)
```

Encotrar area total

```python
area1 = quad(lambda x: f_x(x) - l_tan(x, root_1.root), root_1.root, min["x"][0])
area2 = quad(lambda x: l_tan(x, root_2.root) - f_x(x), root_2.root, min["x"][0])
area3 = quad(lambda x: l_tan(x, root_2.root) - l_tan(x, root_1.root), root_1.root, root_2.root) # 0

total_area = area1[0] + area2[0] + area3[0]
print(f"Area entre las curvas: {total_area}")
```
